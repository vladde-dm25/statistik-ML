Koden börjar med att generera låtsasdata och träna en linjär regressionsmodell på denna. Därefter sparas den färdigtränade modellen ner som en fil, laddas in igen och används slutligen för att göra prediktioner på de första raderna i datan. Att kunna spara modeller är avgörande för effektiviteten i verkliga projekt eftersom modellträning ofta kräver enormt mycket tid och datorkraft. Genom att spara den färdiga modellen kan vi direkt ladda in den i exempelvis en applikation och få omedelbara svar, helt utan att behöva träna om den från noll varje gång den ska användas.

```python
# Importera nödvendiga funktioner och klasser 
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from joblib import dump, load
```

```python
# Skapa ett syntetiskt dataset med 20 000 observationer (rader) och 3 features (kolumner)
# noise=0.1 lägger till lite brus så att datan inte är helt perfekt linjär
X, y = make_regression(n_samples=20000, n_features=3, noise=0.1)

# Skriv ut dimensionerna för att snabbt dubbelkolla att datan ser rätt ut
print(f"Formen på X: {X.shape}")
print(f"Formen på y: {y.shape}")
```

```python
# 1. Instantiera en tom linjär regressionsmodell 
# 2. Träna (fit) modellen på vår genererade data (X och y)
model = LinearRegression().fit(X, y)

print("Modellen är färdigtränad!")
```

```python
# Spara den tränade modellen 
# Filen kommer att dyka upp i samma mapp som din Jupyter Notebook
dump(model, "linear_model.joblib")

print("Modellen har sparats som 'linear_model.joblib'")
```

```python
# Ladda in modellen från filen och spara den i en ny variabel 
model_loaded = load("linear_model.joblib")

print("Modellen har laddats in och är redo att användas!")
```

```python
# Använd den laddade modellen för att prediktera y-värden baserat på de 5 första raderna i X
predictions = model_loaded.predict(X[:5])

print("Modellens prediktioner för de fem första observationerna:")
print(predictions)
```
