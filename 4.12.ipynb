koden hämtar in nödvändiga verktyg från Python-biblioteken och skapar ett enkelt syntetiskt övningsdataset för att ha något att träna på. På nästa steg koden skapar och tränar en beslutsträdsmodell på det genererade datan, där modellens djup medvetet har begränsats till två nivåer för att den inte ska bli onödigt komplicerad. Till slut används en ritfunktion för att skapa en tydlig och grafiskt tilltalande bild av det färdiga beslutsträdet, komplett med färglagda och avrundade noder samt utskrivna namn på både indatavariabler och svarsklasser så att modellens logic  kan följas.

```python

import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.tree import DecisionTreeClassifier, plot_tree

# 1. Ladda det syntetiska datasetet
X, y = make_moons(n_samples=200, noise=0.4, random_state=42)

# 2. Skapa och träna beslutsträdet (begränsat djup 2 för tydlighet)
tree_clf = DecisionTreeClassifier(max_depth=2, random_state=42)
tree_clf.fit(X, y)

# 3. Rita upp visualiseringen
plt.figure(figsize=(12, 6))

plot_tree(
    tree_clf, 
    filled=True,                  # Fyller noderna med färg baserat på klass
    feature_names=["X1", "X2"],   # Namnger variablerna
    class_names=["0", "1"],       # Namnger klasserna
    rounded=True                  # Gör nodernas hörn rundade för ett mjukare utseende
)

plt.title("Beslutsträd")
plt.show()
```
